# Two Sum

```python
# Code :
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        l=len(nums)
        num={}
        for i in range(l):
            res= target-nums[i]
            if res in num:
                return [num[res], i]
            num[nums[i]]=i
        return []
    
# 🧪 One More Quick Example
# nums = [3, 2, 4]
# target = 6

# Steps:

# i=0 → need 3 → not found → store {3:0}
# i=1 → need 4 → not found → store {3:0, 2:1}
# i=2 → need 2 → found → answer [1,2]

```

# Remove Duplicates from Sorted List

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        res = head 
        while head and head.next :
            if head.val == head.next.val :
                head.next = head.next.next 
            else:
                head = head.next 
        return res 
# Input:
# 1 → 1 → 2 → 3 → 3
# Initial:
# res → 1 → 1 → 2 → 3 → 3
# head ↑
# 🔹 Iteration 1
# head = 1
# head.next = 1

# 👉 Compare:

# 1 == 1 → YES

# 👉 Remove duplicate:

# head.next = head.next.next

# Now:

# 1 → 2 → 3 → 3
```

#  Remove Duplicates from Sorted Array II

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        k=2
        n=len(nums)
        for i in range(2,n):
            if nums[i] != nums[k-2]:
                nums[k]= nums[i]
                k=k+1
        return k

# Index:  0   1   2   3   4   5
# nums : [1 | 1 | 1 | 2 | 2 | 3]

# k = 2  (first 2 elements are always valid ✅)

# Iteration 1 (i = 2)
# nums[i] = 1
# Compare with nums[k-2] = nums[0] = 1
# 1 == 1 → ❌ Skip (already 2 duplicates)

# Iteration 2 (i = 3)
# nums[i] = 2
# Compare with nums[k-2] = nums[0] = 1
# 2 != 1 → ✅ Keep
# Before:
# [1 | 1] | 1 | 2 | 2 | 3

# After placing:
# [1 | 1 | 2] | 2 | 2 | 3
# k = 3


# Iteration 3 (i = 4)
# nums[i] = 2
# Compare with nums[k-2] = nums[1] = 1
# 2 != 1 → ✅ Keep
# 👉 Place at index k = 3
# Visual update:
# Before:
# [1 | 1 | 2] | 2 | 2 | 3

# After:
# [1 | 1 | 2 | 2] | 2 | 3
#                 ↑k moves
# k = 4


# Iteration 4 (i = 5)
# nums[i] = 3
# Compare with nums[k-2] = nums[2] = 2
# 3 != 2 → ✅ Keep
# 👉 Place at index k = 4
# Visual update:
# Before:
# [1 | 1 | 2 | 2] | 2 | 3

# After:
# [1 | 1 | 2 | 2 | 3] | 3
#                     ↑k moves
# k = 5
```
